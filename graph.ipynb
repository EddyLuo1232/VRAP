```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Set style and color palette
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans', 'Liberation Sans']
plt.rcParams['axes.unicode_minus'] = False

# Data configuration - Performance comparison across different methods
methods = ['MCTS', 'Genetic Algorithm', 'Few Shots']
scenario_a_scores = [78.58, 78.08, 76.02]  # BI-DIRECTIONAL performance
scenario_b_scores = [77.84, 76.38, 74.06]  # UNIDIRECTIONAL performance

# Create figure with enhanced styling
fig, ax = plt.subplots(figsize=(8, 8))

# Configure bar positions and styling
x_positions = np.arange(len(methods))
bar_width = 0.35
bar_spacing = 0.05

# Define color scheme based on the reference image
colors = {
    'scenario_a': '#FFA726',  # Orange/amber color for BI-DIRECTIONAL
    'scenario_b': '#42A5F5',  # Light blue color for UNIDIRECTIONAL
    'edge': '#37474F'         # Dark gray for edges
}

# Create grouped bar chart without borders
bars_a = ax.bar(x_positions - bar_width/2 - bar_spacing/2, scenario_a_scores, 
                bar_width, label='BI-DIRECTIONAL', color=colors['scenario_a'], 
                alpha=0.85, edgecolor='none', linewidth=0,
                capsize=5)

bars_b = ax.bar(x_positions + bar_width/2 + bar_spacing/2, scenario_b_scores, 
                bar_width, label='UNIDIRECTIONAL', color=colors['scenario_b'], 
                alpha=0.85, edgecolor='none', linewidth=0,
                capsize=5)

# Add value labels above bars
def add_value_labels(bars, values, color='black'):
    for bar, value in zip(bars, values):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 0.2,
                f'{value:.1f}%',
                ha='center', va='bottom', fontsize=14, 
                fontweight='bold', color=color)

add_value_labels(bars_a, scenario_a_scores)
add_value_labels(bars_b, scenario_b_scores)

# Enhanced title and labels
ax.set_title('Pass@K Performance Comparison Across Methods', 
             fontsize=18, fontweight='bold', pad=25, color='#2C3E50')
ax.set_xlabel('Optimization Methods', fontsize=14, fontweight='bold', color='#34495E')
ax.set_ylabel('Pass@K Score (%)', fontsize=14, fontweight='bold', color='#34495E')

# Configure x-axis
ax.set_xticks(x_positions)
ax.set_xticklabels(methods, fontsize=12, fontweight='bold')

# Make y-axis tick labels bold
for label in ax.get_yticklabels():
    label.set_fontweight('bold')

# Set y-axis range with padding
all_scores = scenario_a_scores + scenario_b_scores
y_min = min(all_scores) - 5
y_max = max(all_scores) + 3
ax.set_ylim(y_min, y_max)

# Clean legend without border
legend = ax.legend(fontsize=12, loc='upper right', frameon=False, 
                  fancybox=False, shadow=False, borderpad=1)
# Make legend text bold
for text in legend.get_texts():
    text.set_fontweight('bold')

# Improved grid
ax.grid(axis='y', alpha=0.4, linestyle='-', linewidth=0.5, color='#BDC3C7')
ax.set_axisbelow(True)

# Professional styling
for spine in ax.spines.values():
    spine.set_linewidth(1.2)
    spine.set_color('#34495E')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set background
ax.set_facecolor('#FDFEFE')
fig.patch.set_facecolor('white')

# Remove annotation boxes - keep only the legend

# Optimize layout
plt.tight_layout()

# Display the chart
plt.show()

# Clean chart with legend only
```

```python
# 
import matplotlib.pyplot as plt
import numpy as np

# 
def setup_chart_style():
    """"""
    plt.style.use('default')
    plt.rcParams['font.family'] = 'sans-serif'
    plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans', 'Liberation Sans']
    
    # 
    colors = {
        'primary': '#FF6B6B',     # 
        'secondary': '#4ECDC4',   #   
        'accent': '#45B7D1',      # 
        'warning': '#FFA726',     # 
        'edge': '#34495E',        # 
        'text': '#2C3E50',        # 
        'grid': '#BDC3C7'         # 
    }
    return colors

def apply_chart_styling(ax, title, xlabel, ylabel):
    """"""
    colors = setup_chart_style()
    
    # 
    ax.set_title(title, fontsize=18, fontweight='bold', pad=25, color=colors['text'])
    ax.set_xlabel(xlabel, fontsize=14, fontweight='bold', color=colors['edge'])
    ax.set_ylabel(ylabel, fontsize=14, fontweight='bold', color=colors['edge'])
    
    # 
    for label in ax.get_xticklabels() + ax.get_yticklabels():
        label.set_fontweight('bold')
    
    # 
    ax.grid(axis='y', alpha=0.4, linestyle='-', linewidth=0.5, color=colors['grid'])
    ax.set_axisbelow(True)
    
    # 
    for spine in ax.spines.values():
        spine.set_linewidth(1.2)
        spine.set_color(colors['edge'])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    # 
    ax.set_facecolor('#FDFEFE')

# 1: 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
colors = setup_chart_style()

# 
categories = ['Algorithm A', 'Algorithm B', 'Algorithm C', 'Algorithm D']
values = [92.5, 87.3, 84.1, 79.8]

bars = ax1.barh(categories, values, color=colors['primary'], alpha=0.85, 
                edgecolor=colors['edge'], linewidth=1.2)

# 
for i, (bar, value) in enumerate(zip(bars, values)):
    ax1.text(value + 1, i, f'{value:.1f}%', 
             va='center', ha='left', fontsize=12, fontweight='bold')

apply_chart_styling(ax1, 'Horizontal Bar Chart Example', 'Accuracy (%)', 'Algorithms')

# 2: 
methods = ['Method 1', 'Method 2', 'Method 3']
precision = [85, 78, 82]
recall = [80, 75, 79]
f1_score = [82, 76, 80]

x = np.arange(len(methods))
width = 0.6

p1 = ax2.bar(x, precision, width, label='Precision', color=colors['primary'], alpha=0.85)
p2 = ax2.bar(x, recall, width, bottom=precision, label='Recall', color=colors['secondary'], alpha=0.85)
p3 = ax2.bar(x, f1_score, width, bottom=np.array(precision) + np.array(recall), 
             label='F1-Score', color=colors['accent'], alpha=0.85)

ax2.set_xticks(x)
ax2.set_xticklabels(methods, fontsize=12, fontweight='bold')

# 
legend = ax2.legend(fontsize=11, loc='upper right', frameon=False, 
                   fancybox=False, shadow=False)
for text in legend.get_texts():
    text.set_fontweight('bold')

apply_chart_styling(ax2, 'Stacked Bar Chart Example', 'Methods', 'Scores')

plt.tight_layout()
plt.show()

print(" ")
```
